# Runge2D Plots

This notebook shows the plots for all functions and errors in Section 3 of Project 3.

```python
%pylab inline
```

```python
a = loadtxt("avals.txt")
b = loadtxt("bvals.txt")
runge = loadtxt("runge.txt")
p6_uni = loadtxt("p6.uni.txt")
p24_uni = loadtxt("p24.uni.txt")
p6_Cheb = loadtxt("p6.Cheb.txt")
p24_Cheb = loadtxt("p24.Cheb.txt")
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(b, a)
surf = ax.plot_surface(X, Y, runge, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$f(x,y)$')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, p6_uni, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_6(x,y)$ with Uniform Nodes')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, p24_uni, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_{24}(x,y)$ with Uniform Nodes')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, p6_Cheb, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_6(x,y)$ with Chebyshev Nodes')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, p24_Cheb, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$p_{24}(x,y)$ with Chebyshev Nodes')
```

Error computations:

```python
err6u = abs(runge - p6_uni)
err24u = abs(runge - p24_uni)
err6c = abs(runge - p6_Cheb)
err24c = abs(runge - p24_Cheb)
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, err6u, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$err_{6,uni} = |f(x,y) - p_{6,uni}(x,y)|$')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, err24u, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$err_{24,uni} = |f(x,y) - p_{24,uni}(x,y)|$')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, err6c, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$err_{6,Cheb} = |f(x,y) - p_{6,Cheb}(x,y)|$')
```

```python
fig = figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, err24c, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$err_{24,Cheb} = |f(x,y) - p_{24,Cheb}(x,y)|$')
```
