```python
n = loadtxt('n.txt')
h = loadtxt('h.txt')
r = loadtxt('r.txt')
R = loadtxt('R.txt')
```

```python
semilogy(n, r, 'b-', label='r')
semilogy(n, R, 'r--', label='R')
xlabel('n')
ylabel('Error')
title('Relative and Maximum Errors of Forward Difference Approximation')
legend()
```

```python
loglog(h, r, 'b-', label='r')
loglog(h, R, 'r--', label='R')
xlabel('h')
ylabel('Error')
title('Relative and Maximum Errors of Forward Difference Approximation')
legend()
```
