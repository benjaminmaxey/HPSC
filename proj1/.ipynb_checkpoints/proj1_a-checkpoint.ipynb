```python
z = loadtxt('z.txt')

f = loadtxt('f.txt')

p4 = loadtxt('p4.txt')
p8 = loadtxt('p8.txt')
p12 = loadtxt('p12.txt')

err4 = loadtxt('err4.txt')
err8 = loadtxt('err8.txt')
err12 = loadtxt('err12.txt')
```

```python
plot(z,f,z,p4,z,p8,z,p12)
xlabel('z')
ylabel('Value of polynomial')
title('Approximations of $e^x$ by Taylor Polynomials')
legend(('$f(x) = e^x$','$p_4(x)$','$p_8(x)$','$p_{12}(x)$'))
```

```python
plot(z,err4,z,err8,z,err12)
xlabel('z')
ylabel('Error ($10^{-8})')
title('Error in Approximations')
legend(('$err_4(x)$','$err_8(x)$','$err_{12}(x)$'))
```
